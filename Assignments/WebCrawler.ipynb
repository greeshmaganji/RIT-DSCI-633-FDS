```python
#Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import selenium
from selenium.webdriver.support.ui import Select
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import re
```

This code runs a we scraper over every congressperson on every bill. After it gets the HTML elements, it stores them in numpy arrays.

```python
#loop over all bills (the first 2 are not actually bills)
#THIS WILL TAKE A LOOOOOONNGGG TIME!!!

partyLineAll = []
stateAll = []
congressPeopleAll = []
votesAll = []

for bill in range(3,313):
    #Install chrome driver
    driver = webdriver.Chrome(ChromeDriverManager().install())
    #Go to congressional voting database
    driver.get('https://www.govtrack.us/congress/votes/117-2021/h'+str(bill))
    #Get the Congress person, party, vote, and bill #
    congressPeople = []
    votes = []
    partyLine = []
    state = []
    for j in range(1,4):
        for i in range(1,200):
            try:
                partyVote = driver.find_element_by_xpath('//*[@id="vote-details-all"]/div/table['+str(j)+']/tbody/tr['+str(i)+']')
                votes.append(partyVote.get_attribute("voter_group_0"))
                state.append(partyVote.get_attribute("voter_group_1"))
                partyLine.append(partyVote.get_attribute("voter_group_2"))
                names = driver.find_element_by_xpath('//*[@id="vote-details-all"]/div/table['+str(j)+']/tbody/tr['+str(i)+']/td[4]/a')
                congressPeople.append("".join([i for i in names.get_attribute('href')[41:] if not i.isdigit()])[:-1])
                
            except:
                pass
    votesAll.append(votes)
    congressPeopleAll.append(congressPeople)
    partyLineAll.append(partyLine)
    stateAll.append(state)
    driver.close()
            
```

```python
"""
Save data as a CSV. It is formated as an array of arrays.
Array 0: Name of all congress people who vote on a bill
Array 1: State the congress people are from
Array 2: party of congress people
Array 3: How they voted on a give bill

!!! This is a ragged array because not all congress people vote on every bill !!!

"""

data = np.array([congressPeopleAll,stateAll,partyLineAll,votesAll],dtype=object)
np.savetxt("data.csv",data, fmt='%s')
```
